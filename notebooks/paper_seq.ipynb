```python
def plot_sentences_bar(ax, sampled, tauh=.2, taub=.2, nrefs=5):
    h = .1
    hammings = []
    bleus = []
    sentences = []
    for e, c in enumerate(sampled):
        if e >= nrefs:
            break
        sentences.append(r'\textbf{%s}' % c)
        hammings.append(0)
        bleus.append(1)
        for sabh in sampled[c]:
            sentences.append(r'%s' % ' '.join(highlight(sabh['sample'],
                                                        sabh['alter'])))
            hammings.append(sabh['hamming'])
            bleus.append(sabh['bleu'])
    if 1:
        hammings = np.array([exp(-d/tauh) for d in hammings])
        hammings /= np.sum(hammings)
        bleus = np.array([exp((bleu-1)/taub) for bleu in bleus])
        bleus /= np.sum(bleus)
            
    ys = h/4 * np.arange(2 * len(sentences))
    ax.barh(ys[1::2], hammings[::-1], height=h/4,
            align='edge', color=COLORS[2], alpha=.7, label='$r_{hamming}$')
    ax.barh(ys[::2], bleus[::-1], height=h/4,
            align='edge', color=COLORS[3], alpha=.7, label='$r_{bleu}$')
    ax.set_yticks(ys[1::2])
    ax.set_yticklabels(sentences[::-1])
#     ax.set_xlim([1e-5, 1])
    ax.set_ylim([min(ys), max(ys)+h/4])
    ax.set_xscale("log", nonposx='clip')
    ax.legend(bbox_to_anchor=(.3, -.2), loc=3,
              ncol=2, borderaxespad=0.)
```

```python
def plot_sentences_2d(ax, X2, sampled, refs, pick='hamming', tolv=5, tolh=5, cut=2,
                      tauh=.2, taub=.2):
    keys = list(sampled)
    selected = [keys[e] for e in refs]
    for e, c in enumerate(selected):
        hammings = []
        bleus = []
        sentences = []
        sentences.append(fold(c, 4, bf=True))
        hammings.append(0)
        bleus.append(1)
        NS = len(sampled[c]) + 1
        for sabh in sampled[c]:
            if sabh['hamming']:
                sentences.append(r'%s' % fold(highlight(sabh['sample'],
                                                        sabh['alter']),
                                             4))
                hammings.append(sabh['hamming'])
                bleus.append(sabh['bleu'])
        hammings = np.array([exp(-d/tauh) for d in hammings])
        hammings /= np.sum(hammings)
        bleus = np.array([exp((bleu-1)/taub) for bleu in bleus])
        bleus /= np.sum(bleus)
        V = np.arange(len(sentences))
        X2ref = X2[refs[e]*NS:(refs[e]+1)*NS] 
        if pick=='hamming':
            p = hammings
            samples = np.random.choice(V, p=p, size=(2000,))
            cmap = 'Reds'
            label = "$r_\\tau$, $r:$=Hamming"
        elif pick=="bleu":
            p = bleus
            samples = np.random.choice(V, p=p, size=(1000,))
            cmap = 'Blues'
            label = "$r_\\tau$, $r:=$BLEU-4"
        elif pick=="importance":
            importance = bleus/hammings
            p = importance / np.sum(importance)
            samples = np.random.choice(V, p=p, size=(2000,))
            cmap = 'Greens'
            label = "$r:=$BLEU-4 / $q:=$Hamming"
#         cticks = np.linspace(np.min(p), np.max(p), 5)
        xs = X2ref[samples, 0]
        ys = X2ref[samples, 1]
        cbar_kws={"label": label,
                  "orientation": "horizontal",
                  "pad": .05,
                  "format": "%.1e",
                  'shrink': 0.8}
        kde_kws = {'cut': cut,
                   'shade': True, 'shade_lowest': False,
                   'cbar': False,
                   'cmap': cmap, 'alpha': 0.6,
                   }   
        sns.kdeplot(xs, ys, ax=ax,
                    **kde_kws,
                    cbar_kws=cbar_kws)
        for i in V:
            ax.text(X2ref[i,0], X2ref[i,1], sentences[i], fontsize=14, color="k")
            
        xmin = np.min(X2ref[V,0])
        xmax = np.max(X2ref[V,0])
        xmin -= (xmax-xmin)/tolh
        xmax += (xmax-xmin)/tolh
        ax.set_xlim(xmin, xmax)
        ymin = np.min(X2ref[V,1])
        ymax =  np.max(X2ref[V,1])
        ymin -= (ymax-ymin)/tolv
        ymax += (ymax-ymin)/tolv
        ax.set_ylim([ymin,ymax])
        ax.set_xticklabels([])
        ax.set_yticklabels([])        
```
