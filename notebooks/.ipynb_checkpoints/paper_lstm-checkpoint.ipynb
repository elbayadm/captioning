```python
def plot_pt_heatmap_multi_packed(pt, ids=IDS, id1=0, id2=0, add_cbar=False, indcap=None):
    h =  pt[id1][id2]
    if indcap is None:
        cap = caps_val[ids[id1][id2 // 5]['id']][id2 % 5] + ['EOS']
        indcap = [MATCH.get(w, 9487) for w in cap]
    else:
        try:
            length = list(indcap).index(MATCH['EOS']) + 1
            print('length:', length)
            indcap = indcap[:length]
        except:
            pass
        cap = [WORDS[i] for i in indcap]
        
    h = h[:len(indcap), :]
    columns = []
    for e, row in enumerate(h):
        annot = [""] * len(row)
        candidates = []
        th = np.percentile(row, 99.97)
        highp = np.nonzero(row > th)[0]
        # highp = row.argsort()[-2:][::-1]
        found = False
        for k in highp:
            annot[k] = "%.2f" % row[k]
            annot[k] = WORDS[k]
            candidates.append(k)
            if k == indcap[e]:
                annot[k] = annot[k]+"*"
                found = True
        if not found:
#             print('true token missing!')
            # annot[indcap[e]] = "%.2f*" % row[indcap[e]]
            annot[indcap[e]] = "%s*" % cap[e]
            candidates.append(indcap[e])
        columns.append((candidates, row[candidates], np.array(annot)[candidates]))

    vmin = np.min([np.min(c[1]) for c in columns])
    vmax = np.max([np.max(c[1]) for c in columns])
    print('min', vmin, 'max', vmax)
    figh = 5  * .5
    figw = len(columns) * 1.8
    gridspec_kw = {"width_ratios": [1] * (len(columns) - 1) + [1.2]}
    heatmapkws = dict(square=False,
                      robust=True,
                      cmap="Blues",
                      linewidths=1.0,
                      vmin=vmin, vmax=vmax,
                      fmt='') 
    cbar_kws=dict(label="$p_\\theta(y)$",
                  format="%.2f",
                  orientation="horizontal"
                  )
    annot_kws={'fontsize': 16}
    sns.set()
    sns.set_style('whitegrid', rc={"axes.edgecolor": 'darkgray'})
    sns.set_context('paper', font_scale=2.3, rc={"lines.linewidth": 2,
                                                 "lines.markersize": 8})
    fig, axes = plt.subplots(1, len(columns), figsize=(figw, figh), gridspec_kw=gridspec_kw)
    for e, (cand, proba, annot) in enumerate(columns):
        yticklabels = [WORDS[k] for k in cand]
        X = proba.T[:, np.newaxis]
        TX = annot.T[:, np.newaxis]
        sns.heatmap(X,
                    ax=axes[e],
                    annot=TX, annot_kws=annot_kws,
                    cbar= e==len(columns)-1,
                    yticklabels=[],
                    xticklabels=['t=1' if not e else e+1],
                    **heatmapkws)
#         axes[e].set_xticks(np.arange(lengths[e])+.5)
#         axes[e].set_yticklabels(yticklabels, rotation='horizontal', fontsize=7)
#     axes[0].annotate('$t$', xy=(0, 0), ha='left', va='top', xycoords='axes fraction', fontsize=20)
    if add_cbar:
        cax = fig.add_axes([0.2,.9,0.6,0.02])
        sm = matplotlib.cm.ScalarMappable(cmap=cmap, norm=norm)
        sm.set_array([])
        fig.colorbar(sm, cax=cax, **cbar_kws)
    plt.subplots_adjust(hspace=0.01, wspace=.1)
```

```python
def plot_pt_decode(pt, sampled, ids, id1=0, id2=0):
    h =  pt[id1][id2]
    indcap = sampled[id1][id2]
    eos = np.where(indcap==MATCH['EOS'])[0]
    if len(eos):
        indcap = indcap[:eos[0]+1]
    cap = [WORDS[i] for i in indcap]
    print('caption', cap)
    h = h[:len(indcap), :]
    columns = []
    for e, row in enumerate(h):
        annot = [""] * len(row)
        candidates = []
        th = np.percentile(row, 99.97)
        highp = np.nonzero(row > th)[0]
        # highp = row.argsort()[-2:][::-1]
        found = False
        for k in highp:
            annot[k] = "%.2f" % row[k]
            annot[k] = WORDS[k]
            candidates.append(k)
            if k == indcap[e]:
                annot[k] = annot[k]+"*"
                found = True
        if not found:
            # print('true token missing!')
            # annot[indcap[e]] = "%.2f*" % row[indcap[e]]
            annot[indcap[e]] = "%s*" % cap[e]
            candidates.append(indcap[e])
        columns.append((candidates, row[candidates], np.array(annot)[candidates]))

    vmin = np.min([np.min(c[1]) for c in columns])
    vmax = np.max([np.max(c[1]) for c in columns])
    print('min', vmin, 'max', vmax)
    figh = 5  * .5
    figw = len(columns) * 1.5 + figh
    gridspec_kw = {"width_ratios":[3] + [1] * (len(columns) - 1) + [1.2]}
    heatmapkws = dict(square=False,
                      robust=True,
                      cmap="Blues",
                      linewidths=1.0,
                      vmin=vmin, vmax=vmax,
                      fmt='') 
    cbar_kws=dict(label="$p_\\theta(y)$",
                  format="%.2f",
                  orientation="horizontal"
                  )
    annot_kws={'fontsize': 16}
    sns.set()
    sns.set_style('whitegrid', rc={"axes.edgecolor": 'darkgray'})
    sns.set_context('paper', font_scale=2.3, rc={"lines.linewidth": 2,
                                                 "lines.markersize": 8})
    fig, axes = plt.subplots(1, 1+ len(columns), figsize=(figw, figh), gridspec_kw=gridspec_kw)
    for e, (cand, proba, annot) in enumerate(columns):
        yticklabels = [WORDS[k] for k in cand]
        X = proba.T[:, np.newaxis]
        TX = annot.T[:, np.newaxis]
        sns.heatmap(X,
                    ax=axes[e+1],
                    annot=TX, annot_kws=annot_kws,
                    cbar= e==len(columns)-1,
                    yticklabels=[],
                    xticklabels=['t=1' if not e else e+1],
                    **heatmapkws)
        # axes[e].set_xticks(np.arange(lengths[e])+.5)
        # axes[e].set_yticklabels(yticklabels, rotation='horizontal', fontsize=7)
    # axes[0].annotate('$t$', xy=(0, 0), ha='left', va='top', xycoords='axes fraction', fontsize=20)
    im = mpimg.imread("../data/coco/images/" + ids[id1][id2]['file_path'])
    axes[0].imshow(im)
    axes[0].axis("off")
    plt.subplots_adjust(hspace=0.01, wspace=.1)
```

```python
# TODO: limit vocab to Vsub or Vsub union neighbors(Vsub, n)
def plot_pt_heatmap(ax, pt, ids=IDS, id1=0, id2s=[0],
                    dyn=False, rotate=False, robust=False, cbar=True):
    # id1  batch index
    # id2 : caption index in batch (the image index would be id2//5)
    # sub is either own, ref or batch
    # assert all elements id2s share the same quotient to 5
    hs = []
    annots = []
    lengths = []
    for id2 in id2s:
        h =  pt[id1][id2]
        cap = caps_val[ids[id1][id2 // 5]['id']][id2 % 5] + ['EOS']
        indcap = [MATCH.get(w, 9487) for w in cap]
        h = h[:len(indcap), :]
        lengths.append(len(indcap))
        annot = []
        for row in h:
            annot.append([""] * len(row))

        if not dyn:
            for e, (k ,w) in enumerate(zip(indcap, cap)):
                annot[e][k] = w
        else:
            for e, row in enumerate(h):
                highp = row.argsort()[-3:][::-1]
                for k in highp:
                    annot[e][k] = WORDS[k]
                    try:
                        if annot[e][k] == cap[e]:
                            annot[e][k] = r"\underline{%s}" % annot[e][k]
                    except:
                        pass
        annot = np.array(annot)
        hs.append(h)
        annots.append(annot)
    hs = np.vstack(hs)
    annots = np.vstack(annots)
    selection = np.unique([MATCH.get(w, 9487)
                           for cap in caps_val[ids[id1][id2 // 5]['id']]
                           for w in cap + ['EOS'] ])
#     print('Vocab', selection)
    hs = hs[:, selection]
    annots = annots[:, selection]
#     print('annotations:', annots.shape)
#     print(annots)
#     print('hs:', hs.shape)
#     print(hs)


    if rotate:
        hs = hs.T
        annots = annots.T
#     h = h[::-1, :]
#     annot = annot[::-1, :]
    sns.heatmap(hs, ax=ax,
                cmap="Blues",
#                 vmin=.1, vmax=.95,
                annot=annots, fmt='',
                robust=robust,
                cbar=cbar,
                annot_kws={'fontsize': 10},
                cbar_kws={"label": "$p_\\theta(y)$",
                          "orientation": "horizontal",
                          "pad": .02,
                          "format": "%.2f",
                          'shrink': 0.7}
               )
    ax.set_yticks(np.arange(len(hs))+.5)
    tlabels = np.hstack([np.arange(l) for l in lengths])
    ax.set_yticklabels(tlabels)
    ax.set_xticklabels([])
```

```python
def plot_pt_heatmap_multi_hor(pt, ids=IDS, id1=0, id2s=[0]):
    hs = []
    annots = []
    lengths = []
    for id2 in id2s:
        h =  pt[id1][id2]
        cap = caps_val[ids[id1][id2 // 5]['id']][id2 % 5] + ['EOS']
        indcap = [MATCH.get(w, 9487) for w in cap]
        h = h[:len(indcap), :]
        lengths.append(len(indcap))
        annot = []
        for row in h:
            annot.append([""] * len(row))
        candidates = []
        found = False
        for e, row in enumerate(h):
            th = np.percentile(row, 99.98)
            highp = np.nonzero(row > th)[0]
            highp = row.argsort()[-2:][::-1]
            for k in highp:
                # annot[e][k] = WORDS[k]
                annot[e][k] = "%.2f" % row[k]
                candidates.append(k)
                if k == indcap[e]:
                    # annot[e][k] = r"\underline{%s}" % annot[e][k]
                    annot[e][k] = annot[e][k]+"*"
                    found = True
            if not found:
                print('true token missing!')
                annot[e][indcap[e]] = r"\bf %s" % cap[e]
        annot = np.array(annot)
        hs.append(h)
        annots.append(annot)
    selection = np.unique([MATCH.get(w, 9487)
                           for cap in caps_val[ids[id1][id2 // 5]['id']]
                           for w in cap + ['EOS'] ] + candidates)
    hs = [h[:, selection] for h in hs]
    annots = [annot[:, selection] for annot in annots]
    # vmin = np.min([np.min(h) for h in hs])
    # vmax = np.max([np.max(h) for h in hs])
    vmin = np.min([np.percentile(h, 97) for h in hs])
    vmax = np.max([np.percentile(h, 99.99) for h in hs])
    print('vmin:', vmin, 'vmax:', vmax)
    cmap = plt.cm.Blues
    norm = matplotlib.colors.Normalize(vmin= vmin, vmax=vmax)
    figw = len(selection)  * .68
    figh = sum(lengths) * .4
    # figh = figw / 2
    gridspec_kw = {"height_ratios": lengths}
    heatmapkws = dict(square=False,
                      robust=True,
                      cmap=cmap,
                      linewidths=1.0,
                      vmin=vmin, vmax=vmax,
                      fmt='') 
    cbar_kws=dict(label="$p_\\theta(y)$",
                  # orientation="horizontal",
                  # pad= .15,
                  format="%.2f",
                  # shrink=0.7
                  )
    annot_kws={'fontsize': 16}
    sns.set()
    sns.set_style('whitegrid', rc={"axes.edgecolor": 'darkgray'})
    sns.set_context('paper', font_scale=2.3, rc={"lines.linewidth": 2,
                                                 "lines.markersize": 8})
    fig, axes = plt.subplots(len(lengths), 1, figsize=(figw, figh), gridspec_kw=gridspec_kw)
    # left = 0.07; right=0.8
    # bottom = 0.1; top = 0.9
    # plt.subplots_adjust(left=left, right=right,
    #                    bottom=bottom, top=top,
    #                    wspace=0.1, hspace=0.1*asp )
    for e, (h, annot) in enumerate(zip(hs, annots)):
        if e == len(hs) - 1:
            cbar=False
            xticklabels = [WORDS[k] for k in selection]
        else:
            cbar=False
            xticklabels = []
        sns.heatmap(h, ax=axes[e],
                    annot=annot, annot_kws=annot_kws,
                    cbar=cbar, cbar_kws=cbar_kws,
                    xticklabels=xticklabels,
                    yticklabels=0,
                    **heatmapkws)
        axes[e].set_yticks(np.arange(lengths[e])+.5)
        axes[e].set_yticklabels(['t=%02d'%t for t in 1 + np.arange(lengths[e])], rotation='horizontal')
    cax = fig.add_axes([0.92,0.2,0.02,0.6])
    sm = matplotlib.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    fig.colorbar(sm, cax=cax, **cbar_kws)
```

```python
def plot_pt_heatmap_multi(pt, ids=IDS, id1=0, id2s=[0], add_cbar=False):
    hs = []
    annots = []
    lengths = []
    for id2 in id2s:
        h =  pt[id1][id2]
        cap = caps_val[ids[id1][id2 // 5]['id']][id2 % 5] + ['EOS']
        indcap = [MATCH.get(w, 9487) for w in cap]
        h = h[:len(indcap), :]
        lengths.append(len(indcap))
        annot = []
        for row in h:
            annot.append([""] * len(row))
        candidates = []
        for e, row in enumerate(h):
            th = np.percentile(row, 99.97)
            highp = np.nonzero(row > th)[0]
            # highp = row.argsort()[-2:][::-1]
            found = False
            for k in highp:
                annot[e][k] = "%.2f" % row[k]
                candidates.append(k)
                if k == indcap[e]:
                    annot[e][k] = annot[e][k]+"*"
                    found = True
            if not found:
                print('true token missing!')
                annot[e][indcap[e]] = r"\bf %s" % cap[e]
                candidates.append(indcap[e])
        annot = np.array(annot)
        hs.append(h)
        annots.append(annot)
    selection = np.unique([MATCH.get(w, 9487)
                           for cap in caps_val[ids[id1][id2 // 5]['id']]
                           for w in cap + ['EOS'] ] + candidates)
    hs = [h[:, selection] for h in hs]
    annots = [annot[:, selection] for annot in annots]
    # vmin = np.min([np.min(h) for h in hs])
    # vmax = np.max([np.max(h) for h in hs])
    vmin = np.min([np.percentile(h, 97) for h in hs])
    vmax = np.max([np.percentile(h, 99.99) for h in hs])
    print('vmin:', vmin, 'vmax:', vmax)
    cmap = plt.cm.Blues
    norm = matplotlib.colors.Normalize(vmin= vmin, vmax=vmax)
    figh = len(selection)  * .5
    figw = sum(lengths) * .68
    # figh = figw / 2
    gridspec_kw = {"width_ratios": lengths}
    heatmapkws = dict(square=False,
                      robust=True,
                      cmap=cmap,
                      linewidths=1.0,
                      vmin=vmin, vmax=vmax,
                      fmt='') 
    cbar_kws=dict(label="$p_\\theta(y)$",
                  format="%.2f",
                  orientation="horizontal"
                  )
    annot_kws={'fontsize': 16}
    sns.set()
    sns.set_style('whitegrid', rc={"axes.edgecolor": 'darkgray'})
    sns.set_context('paper', font_scale=2.3, rc={"lines.linewidth": 2,
                                                 "lines.markersize": 8})
    fig, axes = plt.subplots(1, len(lengths), figsize=(figw, figh), gridspec_kw=gridspec_kw)
    if not isinstance(axes, list):
        axes = [axes]
    for e, (h, annot) in enumerate(zip(hs, annots)):
        if not e:
            yticklabels = [WORDS[k] for k in selection]
        else:
            yticklabels = []
        sns.heatmap(h.T, ax=axes[e],
                    annot=annot.T, annot_kws=annot_kws,
                    cbar=False,
                    yticklabels=yticklabels,
                    xticklabels=0,
                    **heatmapkws)
        axes[e].set_xticks(np.arange(lengths[e])+.5)
        axes[e].set_xticklabels(['%d'%t for t in 1 + np.arange(lengths[e])], rotation='horizontal')
    axes[0].annotate('$t$', xy=(0, 0), ha='left', va='top', xycoords='axes fraction', fontsize=20)
    if add_cbar:
        cax = fig.add_axes([0.2,.9,0.6,0.02])
        sm = matplotlib.cm.ScalarMappable(cmap=cmap, norm=norm)
        sm.set_array([])
        fig.colorbar(sm, cax=cax, **cbar_kws)
    plt.subplots_adjust(hspace=0.01, wspace=.015)
```
